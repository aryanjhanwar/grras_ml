```
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = {
    "gender": ["male", "female", "others", "female", "male"]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

le = LabelEncoder()
df["gender_encoded"] = le.fit_transform(df["gender"])

print("\nAfter Label Encoding")
print(df)
```
